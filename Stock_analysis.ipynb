```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
start = datetime.datetime(2021,1,1)
end = datetime.datetime(2021,12,11)
```

```python
google = web.DataReader("GOOGL",'yahoo',start,end)
```

```python
google.head()
```

# Current Stock prices of google from 2021-1-1 to 2021-12-11

```python
google['Open'].plot(label ="GOOGLE Open Price",figsize=(15,8))
google['Close'].plot(label = "GOOGLE Close Price")
google['High'].plot(label = "GOOGLE High Price")
google['Low'].plot(label = "GOOGLE Low Price")
plt.xlabel('Date')
plt.ylabel('Price of stocks')
plt.title("Google Stock Prices")
plt.legend()
plt.show()
```

# Analysis on stock prices of Tesla,Fordand GM  

```python
start = datetime.datetime(2015,1,1)
end = datetime.datetime(2020,1,1)
```

```python
tesla = web.DataReader("TSLA","yahoo",start,end)
ford = web.DataReader("F","yahoo",start,end)
gm = web.DataReader("GM","yahoo",start,end)
```

```python
#Creating the data into csv
tesla.to_csv('Tesla_stocks_2015-2020.csv')
ford.to_csv('ford_stocks_2015-2020.csv')
gm.to_csv('gm_stocks_2015-2020.csv')
```
